```python
import pandas as pd
data_path = "./data/"
```

```python
domain_fname = "domainsList.csv"
domain_df = pd.read_csv(data_path + domain_fname,names=["domain"])
domain_df
```

```python
fname = "cookieJSON.json"
df = pd.read_json(data_path + fname)
df
```

```python
filtered_by_domain_df = df[df["domain"].str.contains('|'.join(domain_df['domain']), case=False, regex=True)]
percent = len(filtered_by_domain_df)/len(df)*100

print(f'%d different cookies were captured across %d websites' % (len(df), len(domain_df)))
print(f'%.2f%% of cookie domains are first party (the number of cookies from domains that got visited)' % percent)

print(f'\t%.2f%% of cookies had the "secure" flag set' % (len(df[df["secure"]==True]) / len(df)*100))
print(f'\t%.2f%% of cookies had the "hostOnly" flag set' % (len(df[df["hostOnly"]==True]) / len(df)*100))
print(f'\t%.2f%% of cookies had the "httpOnly" flag set' % (len(df[df["httpOnly"]==True]) / len(df)*100))
```
